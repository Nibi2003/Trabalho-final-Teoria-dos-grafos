>> Extraindo features importantes, no domínio do tempo mesmo para realizar o teste

```python
from scipy.stats import skew, kurtosis
from scipy.signal import welch

def hjorth_params(x):
    dx = np.diff(x)
    ddx = np.diff(dx)

    var_x = np.var(x)
    var_dx = np.var(dx)
    var_ddx = np.var(ddx)

    mobility = np.sqrt(var_dx / var_x)
    complexity = np.sqrt(var_ddx / var_dx) / mobility

    return mobility, complexity

def extract_features_from_channel(x, fs=100):
    feats = {}

    # Estatísticas básicas
    feats["mean"] = np.mean(x)
    feats["std"] = np.std(x)
    feats["var"] = np.var(x)
    feats["min"] = np.min(x)
    feats["max"] = np.max(x)
    feats["median"] = np.median(x)
    feats["range"] = np.max(x) - np.min(x)
    feats["iqr"] = np.percentile(x, 75) - np.percentile(x, 25)

    # Energia
    feats["rms"] = np.sqrt(np.mean(x**2))
    feats["energy"] = np.sum(x**2)
    feats["zero_crossings"] = ((x[:-1] * x[1:]) < 0).sum()

    # Distribuição
    feats["skew"] = skew(x)
    feats["kurtosis"] = kurtosis(x)

    # Hjorth
    mobility, complexity = hjorth_params(x)
    feats["hjorth_mobility"] = mobility
    feats["hjorth_complexity"] = complexity

    # Frequência (Welch)
    f, Pxx = welch(x, fs=fs)
    feats["spec_entropy"] = -np.sum(Pxx * np.log(Pxx + 1e-12))
    feats["spec_centroid"] = np.sum(f * Pxx) / np.sum(Pxx)
    feats["spec_spread"] = np.sqrt(np.sum(((f - feats["spec_centroid"])**2) * Pxx) / np.sum(Pxx))

    return feats

def extract_all_features(data):
    """
    data: matriz (61 canais × 2112 tempo)
    retorna: matriz (61 nós × F features)
    """
    feature_list = []
    for i in range(data.shape[0]):
        feats = extract_features_from_channel(data[i])
        feature_list.append(list(feats.values()))

    return np.array(feature_list)

X_global = extract_all_features(data_001)
X_global.shape
```

>> Extraindo features com janelamento

```python
def extract_features_from_window(x):
    feats = {}

    feats["mean"] = np.mean(x)
    feats["std"] = np.std(x)
    feats["rms"] = np.sqrt(np.mean(x**2))
    feats["energy"] = np.sum(x**2)
    feats["skew"] = skew(x)
    feats["kurtosis"] = kurtosis(x)

    return list(feats.values())

def extract_window_features(signal, win_size=100, step=100):
    windows = []
    for start in range(0, len(signal)-win_size+1, step):
        w = signal[start:start+win_size]
        windows.append(extract_features_from_window(w))
    return np.array(windows).flatten()  # concatena features de todas janelas

def extract_all_channels(data, win_size=100):
    feature_list = []
    for ch in range(data.shape[0]):
        feats = extract_window_features(data[ch], win_size)
        feature_list.append(feats)
    return np.array(feature_list)


X_win = extract_all_channels(data_001)
X_win.shape
```

Shuman et al., 2013 “The Emerging Field of Signal Processing on Graphs”
“A construção da matriz de adjacência segue a recomendação apresentada por Shuman et al. (2013), que descrevem o uso de grafos de k-vizinhos (k-NN) como procedimento padrão para modelar relações de similaridade entre sinais antes da aplicação de métodos baseados em convoluções espectrais.”

```python
from sklearn.neighbors import NearestNeighbors

def knn_graph(X, k=5, metric='euclidean'):
    """
    X: matriz (num_nodes × num_features)
    k: número de vizinhos
    retorna: matriz de adjacência (num_nodes × num_nodes)
    """
    n_nodes = X.shape[0]

    nbrs = NearestNeighbors(n_neighbors=k+1, algorithm='auto', metric=metric).fit(X)
    distances, indices = nbrs.kneighbors(X)

    # construir matriz de adjacência
    A = np.zeros((n_nodes, n_nodes))

    for i in range(n_nodes):
        # pular o próprio nó (índice 0 nos vizinhos)
        for j in indices[i][1:]:  
            A[i, j] = 1
            A[j, i] = 1   # grafo não-direcionado

    return A
A_knn = knn_graph(X_win, k=5)
A_knn.shape
```

Representação Esparsa do Grafo
A matriz de adjacência construída a partir do KNN contém majoritariamente zeros e, portanto, é uma matriz esparsa. Para reduzir o custo computacional e o uso de memória, convertemos essa matriz para o formato esparso CSR (Compressed Sparse Row), que armazena apenas as posições com valores diferentes de zero.

Em seguida, utilizamos a função from_scipy_sparse_matrix da biblioteca PyTorch Geometric para transformar essa matriz esparsa no formato edge_index, que representa o grafo como uma lista de arestas. Esse é o formato padrão utilizado pelas redes neurais em grafos (GNNs), pois facilita as operações de message passing.

```python
import torch
from torch_geometric.data import Data
from scipy.sparse import csr_matrix
from torch_geometric.utils import from_scipy_sparse_matrix

# entradas
# X_win: Matriz de features (61, 132)
# A_knn: Matriz de adjacência (61, 61)
# patient_label: Rótulo (0 ou 1)

# Ex: Paciente 001 é Saudável (0)
patient_label = 0 
X_win_tensor = torch.tensor(X_win, dtype=torch.float)

A_sparse = csr_matrix(A_knn)

# 2. Converter a matriz esparsa para o formato PyG (edge_index)
# edge_index: Lista de coordenadas (u, v) das arestas.
# edge_attr: Pesos das arestas (será None, pois A é binária)
edge_index, _ = from_scipy_sparse_matrix(A_sparse)

# 3. Definir o Rótulo (Label)
Y_tensor = torch.tensor([patient_label], dtype=torch.long)

# 4. Criar o Objeto de Dados (Grafo Final)
grafo_paciente = Data(x=X_win_tensor, edge_index=edge_index, y=Y_tensor)

print("\nObjeto PyG do Grafo Criado ")
print(f"Features (x): {grafo_paciente.x.shape}")
print(f"Conectividade (edge_index): {grafo_paciente.edge_index.shape}")
print(f"Rótulo (y): {grafo_paciente.y.item()}")
```

```python
import json

def get_label_from_json(patient_json_path):
    """
    Extrai o rótulo binário rigoroso:
    1 (Parkinson's Disease) ou 0 (Healthy Control).
    Outras condições são descartadas (-1).
    """
    try:
        with open(patient_json_path, "r") as f:
            data = json.load(f)
    except FileNotFoundError:
        return -1
    
    condition_value = data.get("condition", "").lower()
    
    if not condition_value:
        condition_value = data.get("disease_comment", "").lower()
    
    
    if "parkinson's" in condition_value or "pd" == condition_value:
        # Garante que não é "Atypical Parkinsonism" ou "Atypical PD"
        if "atypical" not in condition_value:
            return 1 
        
    if "healthy" in condition_value or "normal" in condition_value:
        return 0
    
    return -1
```

```python
def build_pyg_dataset(base_path):
    """
    Escaneia a pasta de arquivos bin, busca o JSON correspondente,
    e retorna a lista de grafos PyG válidos.
    """
    list_of_graphs = []
    
    # Pastas de dados
    BIN_PATH = os.path.join(base_path, "preprocessed", "movement")
    JSON_PATH = os.path.join(base_path, "patients")
    
    #  Encontra todos os arquivos .bin, pega em ordem alfabética
    bin_filenames = [f for f in os.listdir(BIN_PATH) if f.endswith('_ml.bin')]
    bin_filenames.sort()
    
    print(f"Iniciando processamento de {len(bin_filenames)} arquivos .bin encontrados...")
    cont = 0
    
    # Use 'bin_filename' para iterar sobre a lista de nomes de arquivos
    for bin_filename in bin_filenames:
        # Extrai o ID do paciente (ex: '001')
        subject_id = bin_filename.split('_')[0]
        if(cont < 5):
            print(f"Processando paciente: {subject_id}")
        
        # Cria os caminhos COMPLETOS usando uma NOVA variável: bin_file_path
        bin_file_path = os.path.join(BIN_PATH, bin_filename) 
        json_file_path = os.path.join(JSON_PATH, f"patient_{subject_id}.json")
        
        if not os.path.exists(bin_file_path) or not os.path.exists(json_file_path):
            continue
            
        try:
            # Obter Label (Y)
            patient_label = get_label_from_json(json_file_path)
            # print(f"Label: {patient_label}")
            
            if patient_label == -1:
                continue # Pula se não for Parkinson ou Healthy
                
            # Use bin_file_path aqui
            raw_data = load_subject_bin(bin_file_path)
            
            
            # Extrai Features (X)
            X_win = extract_all_channels(raw_data)
            
            # Construi Adjacência (A)
            A_knn = knn_graph(X_win, k=5)
            
            # Converte para PyTorch Geometric
            X_tensor = torch.tensor(X_win, dtype=torch.float)
            Y_tensor = torch.tensor([patient_label], dtype=torch.long)
            
            A_sparse = csr_matrix(A_knn)
            edge_index, _ = from_scipy_sparse_matrix(A_sparse)
            
            # Criar objeto Data
            grafo_paciente = Data(x=X_tensor, edge_index=edge_index, y=Y_tensor)
            if(cont < 5):
                print("\n Objeto PyG do Grafo Criado -----------")
                print(f"Features (x): {grafo_paciente.x.shape}")
                print(f"Conectividade (edge_index): {grafo_paciente.edge_index.shape}")
                print(f"Rótulo (y): {grafo_paciente.y.item()}")
            
            list_of_graphs.append(grafo_paciente)
            cont = cont + 1
            
        except Exception as e:
            # print(f"Erro ao processar {subject_id}: {e}")
            continue
    
    # É melhor para retornar a lista de grafos
    print(f" Processamento finalizado. Total de {len(list_of_graphs)} grafos válidos criados.")
    return list_of_graphs
```

>2. Padrões de Movimento Únicos de Cada PacientePaciente 001 (554 Arestas) o K-NN pode ter gerado um número menor de conexões bidirecionais únicas.

```python
import random
from torch_geometric.data import DataLoader
from sklearn.model_selection import train_test_split
from collections import Counter

if "dataset" not in locals() or len(dataset) == 0:
    raise ValueError("Erro no dataset")

total_grafos = len(dataset)
print(f"Dataset carregado! Total de grafos: {total_grafos}")

def contar_classes(data_list, titulo=""):
    labels = [d.y.item() for d in data_list]
    cont = Counter(labels)
    if titulo:
        print(f"\n{titulo}")
    for cls, qtd in sorted(cont.items()):
        print(f"  Classe {cls}: {qtd} grafos")
    return cont

labels = [d.y.item() for d in dataset]

# 15% Teste

train_val, test_data = train_test_split(
    dataset,
    test_size=0.15,
    random_state=42,
    shuffle=True,
    stratify=labels,
)

print("\nPrimeira divisão:")
print(f"  Treino+Val: {len(train_val)}")
print(f"  Teste:      {len(test_data)}")
contar_classes(test_data, "Teste")

# 70% treino / 15% val total

train_val_labels = [d.y.item() for d in train_val]

train_data, val_data = train_test_split(
    train_val,
    test_size=0.17647,      # para virar exatamente 15% do total
    random_state=42,
    shuffle=True,
    stratify=train_val_labels,
)

print("\nSegunda divisão:")
print(f"  Treino:     {len(train_data)}  (~70%)")
print(f"  Validação:  {len(val_data)}    (~15%)")
contar_classes(train_data, "Treino")
contar_classes(val_data, "Validação")

#  DataLoaders

BATCH_SIZE = 6

train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
val_loader   = DataLoader(val_data,   batch_size=BATCH_SIZE, shuffle=False)
test_loader  = DataLoader(test_data,  batch_size=BATCH_SIZE, shuffle=False)

print("\n DataLoaders criados")
print(f"  Treino:     {len(train_data)} grafos")
print(f"  Validação:  {len(val_data)} grafos")
print(f"  Teste:      {len(test_data)} grafos")
```

Após a construção dos grafos referentes a cada paciente, realizamos a divisão do dataset em três subconjuntos: treino, validação e teste. O objetivo dessa etapa é garantir uma avaliação adequada do modelo, evitando sobreajuste e permitindo a calibração dos hiperparâmetros.

Primeiro, reservamos 15% do conjunto total exclusivamente para o teste, preservando a distribuição original das classes por meio de divisão estratificada. Em seguida, os 85% restantes foram novamente particionados de forma estratificada, de modo que 70% do dataset total fossem destinados ao treino e 15% à validação.

O uso de estratificação em todas as divisões é essencial para manter o desbalanceamento natural das classes (controle vs. Parkinson), garantindo que o modelo seja treinado e avaliado em distribuições consistentes e representativas do problema real.

Os conjuntos finais ficaram distribuídos aproximadamente da seguinte forma:

70% Treino: utilizado para ajuste dos pesos do modelo.

15% Validação: utilizado para seleção de hiperparâmetros e detecção de sobreajuste.

15% Teste: utilizado exclusivamente como métrica final de desempenho.


```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, global_mean_pool

class GCNClassifier(nn.Module):
    def __init__(self, num_node_features, num_classes):
        super(GCNClassifier, self).__init__()
        
        # O modelo tem 126 features de entrada por nó e 2 classes de saída (0 ou 1)
        
        # Camada 1: GCN Convolution
        # Transforma 126 features em 64
        self.conv1 = GCNConv(num_node_features, 64)
        
        # Camada 2: GCN Convolution
        # Mantém 64 features (captura vizinhança de 2º grau)
        self.conv2 = GCNConv(64, 64)
        
        # Camada 3: Classificação Linear
        # Recebe o vetor de 64 (após o pooling) e o mapeia para 2 classes
        self.lin = nn.Linear(64, num_classes) 

    def forward(self, data):
        # PyG Data: x (features), edge_index (arestas), batch
        x, edge_index, batch = data.x, data.edge_index, data.batch
        
        # 1ª Camada GCN 
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        
        #2ª Camada GCN
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        
        # Pooling Global 
        x = global_mean_pool(x, batch) 
        
        #Classificação
        x = self.lin(x)
        
        return F.log_softmax(x, dim=1)
```

>> Essa parte está mais relacionada com o pytorch

```python
import os
import json
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

from sklearn.metrics import f1_score
from torch_geometric.loader import DataLoader  

NUM_NODE_FEATURES = 126
NUM_CLASSES = 2
EPOCHS = 100
LR = 0.001
BATCH_SIZE = 6

RESULTS_DIR = "./results"
METRICS_DIR = os.path.join(RESULTS_DIR, "metrics")
os.makedirs(METRICS_DIR, exist_ok=True)


def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)


def train_epoch(model, optimizer, criterion, device, loader):
    model.train()
    total_loss = 0.0

    for data in loader:
        data = data.to(device)
        optimizer.zero_grad()
        out = model(data)
        loss = criterion(out, data.y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * data.num_graphs

    return total_loss / len(loader.dataset)


def evaluate(model, criterion, device, loader):
    model.eval()
    total_loss = 0.0
    y_true = []
    y_pred = []
    correct = 0

    with torch.no_grad():
        for data in loader:
            data = data.to(device)
            out = model(data)
            loss = criterion(out, data.y)
            total_loss += loss.item() * data.num_graphs

            pred = out.argmax(dim=1)
            correct += int((pred == data.y).sum())
            y_true.extend(data.y.cpu().tolist())
            y_pred.extend(pred.cpu().tolist())

    avg_loss = total_loss / len(loader.dataset)
    accuracy = correct / len(loader.dataset)
    f1 = f1_score(y_true, y_pred, average="binary")

    return avg_loss, accuracy, f1


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

model = GCNClassifier(NUM_NODE_FEATURES, NUM_CLASSES).to(device)
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr=LR)

print("\n--- Iniciando treinamento GCN ---\n")

history = {
    "epoch": [],
    "train_loss": [],
    "val_loss": [],
    "train_acc": [],
    "val_acc": [],
    "train_f1": [],
    "val_f1": []
}

best_val_f1 = 0.0
best_epoch = 0
best_state = None

for epoch in range(1, EPOCHS + 1):
    train_loss = train_epoch(model, optimizer, criterion, device, train_loader)

    
    _, train_acc, train_f1 = evaluate(model, criterion, device, train_loader)
    val_loss, val_acc, val_f1 = evaluate(model, criterion, device, val_loader)

    history["epoch"].append(epoch)
    history["train_loss"].append(float(train_loss))
    history["val_loss"].append(float(val_loss))
    history["train_acc"].append(float(train_acc))
    history["val_acc"].append(float(val_acc))
    history["train_f1"].append(float(train_f1))
    history["val_f1"].append(float(val_f1))

    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        best_epoch = epoch
        best_state = {
            "epoch": epoch,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict()
        }

    print(
        f"Epoch {epoch:03d} | "
        f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
        f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | "
        f"Val F1: {val_f1:.4f}"
    )

print(f"\nMelhor época (Val F1): {best_epoch} com F1 = {best_val_f1:.4f}")

history_path = os.path.join(METRICS_DIR, "training_history.json")
with open(history_path, "w", encoding="utf-8") as f:
    json.dump(history, f, indent=2, ensure_ascii=False)
print("Histórico salvo em:", history_path)

if best_state is not None:
    best_model_path = os.path.join(RESULTS_DIR, "best_model.pt")
    torch.save(best_state, best_model_path)
    print("Melhor modelo salvo em:", best_model_path)

if best_state is not None:
    model.load_state_dict(best_state["model_state_dict"])

test_loss, test_acc, test_f1 = evaluate(model, criterion, device, test_loader)
print("\n--- Avaliação final no TESTE ---")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Acc:  {test_acc:.4f}")
print(f"Test F1:   {test_f1:.4f}")

final_metrics = {
    "best_val_f1": float(best_val_f1),
    "best_epoch": int(best_epoch),
    "test_loss": float(test_loss),
    "test_acc": float(test_acc),
    "test_f1": float(test_f1)
}

final_metrics_path = os.path.join(METRICS_DIR, "final_metrics.json")
with open(final_metrics_path, "w", encoding="utf-8") as f:
    json.dump(final_metrics, f, indent=2, ensure_ascii=False)
print("Métricas finais salvas em:", final_metrics_path)



epochs = history["epoch"]

# Loss
plt.figure(figsize=(7, 4))
plt.plot(epochs, history["train_loss"], label="Treino")
plt.plot(epochs, history["val_loss"], label="Validação")
plt.xlabel("Época")
plt.ylabel("Loss")
plt.title("Curva de Loss")
plt.legend()
plt.grid(True)
loss_curve_path = os.path.join(METRICS_DIR, "loss_curve.png")
plt.savefig(loss_curve_path, bbox_inches="tight", dpi=300)
plt.close()

# Acurácia
plt.figure(figsize=(7, 4))
plt.plot(epochs, history["train_acc"], label="Treino")
plt.plot(epochs, history["val_acc"], label="Validação")
plt.xlabel("Época")
plt.ylabel("Acurácia")
plt.title("Curva de Acurácia")
plt.legend()
plt.grid(True)
acc_curve_path = os.path.join(METRICS_DIR, "accuracy_curve.png")
plt.savefig(acc_curve_path, bbox_inches="tight", dpi=300)
plt.close()

# F1-score
plt.figure(figsize=(7, 4))
plt.plot(epochs, history["train_f1"], label="Treino")
plt.plot(epochs, history["val_f1"], label="Validação")
plt.xlabel("Época")
plt.ylabel("F1-score")
plt.title("Curva de F1-score")
plt.legend()
plt.grid(True)
f1_curve_path = os.path.join(METRICS_DIR, "f1_curve.png")
plt.savefig(f1_curve_path, bbox_inches="tight", dpi=300)
plt.close()

print("Gráficos salvos em:")
print("  ", loss_curve_path)
print("  ", acc_curve_path)
print("  ", f1_curve_path)
```

```python
from sklearn.metrics import roc_curve, auc
from torch_geometric.data import DataListLoader 

def get_test_scores(model, device, loader):
    model.eval()
    y_true = []
    y_scores = [] # Armazena as probabilidades da classe positiva (1)
    
    for data in loader:
        data = data.to(device)
        with torch.no_grad():
            out = model(data)
            probabilities = torch.exp(out)[:, 1] 
            
        y_true.extend(data.y.tolist())
        y_scores.extend(probabilities.cpu().tolist())
        
    return y_true, y_scores

y_true, y_scores = get_test_scores(model, device, test_loader)

fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR) - Recall')
plt.title('Curva ROC e AUC no Conjunto de Teste')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
# 

print(f"AUC Final no Teste: {roc_auc:.4f}")
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Escolhe um limiar (threshold) pra transformar probabilidade em classe
threshold = 0.5
y_pred = (np.array(y_scores) >= threshold).astype(int)

# Matriz de confusão: linhas = verdade, colunas = predição
cm = confusion_matrix(y_true, y_pred)
tn, fp, fn, tp = cm.ravel()

print("Matriz de Confusão:")
print(cm)
print(f"TN = {tn}, FP = {fp}, FN = {fn}, TP = {tp}")

print("\nRelatório de classificação:")
print(classification_report(y_true, y_pred, digits=4))
```
